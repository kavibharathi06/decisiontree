```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Load data
movie_info = pd.read_csv('Movie_Interests_DecisionTree.csv')

# Split features and target
X = movie_info.drop(columns=['Interest'])
y = movie_info['Interest']

# Train model
movie_model = DecisionTreeClassifier(random_state=42)
movie_model.fit(X, y)

# Create input as DataFrame (IMPORTANT)
test_input = pd.DataFrame(
    [[14, 1], [33, 0]],
    columns=['Age', 'Gender']
)

# Predict
movie_interest = movie_model.predict(test_input)
print(movie_interest)
```
